### 🎯 Resumo visual (super condensado)

*   * Classe → cria → Objeto
    * Objeto → usa → Métodos
    * Objeto → guarda → Atributos

*   * Estimator = qualquer classe com .fit()

*   * Transformadores → fit + transform
    * Modelos → fit + predict

*   * Pipeline = vários estimators → 1 só objeto

---

### 🧠 Scikit-Learn MAPA MENTAL (Versão Compacta)

```text
SCIKIT-LEARN — Mapa Mental (Versão Compacta)

1) Programação Orientada a Objetos
   • Classe → molde (ex.: RandomForestRegressor)
   • Objeto → instância criada (model = ...)
   • Métodos → fit(), transform(), predict()
   • Atributos → feature_importances_, n_estimators

2) Estimator
   • Qualquer classe com .fit()
   • Tipos: modelos, transformadores, pipelines

3) Interface fit / transform / predict
   • fit() → treina
   • transform() → transforma dados
   • fit_transform() → fit + transform
   • predict() → previsões

4) Pipelines
   • Sequência de etapas (imputer → scaler → modelo)
   • Evita leakage, organiza código
   • pipeline.fit() / pipeline.predict()
```

### 🧠 MAPA MENTAL — Arquitetura do scikit‑learn


```text
SCIKIT-LEARN
│
├── 1. Programação Orientada a Objetos
│     │
│     ├── Classe
│     │     • É o molde / projeto / receita
│     │     • Ex.: RandomForestRegressor, SimpleImputer
│     │
│     ├── Objeto (instância)
│     │     • É o modelo real criado a partir da classe
│     │     • Ex.: model = RandomForestRegressor()
│     │
│     ├── Métodos
│     │     • Ações que o objeto executa
│     │     • fit(), transform(), predict(), score()
│     │
│     └── Atributos
│           • Informações armazenadas no objeto
│           • Ex.: model.feature_importances_, model.n_estimators
│
├── 2. Estimator (conceito central)
│     │
│     ├── Definição
│     │     • Qualquer classe que implementa .fit()
│     │     • Pode ser modelo, transformador ou pipeline
│     │
│     ├── Tipos
│     │     • Modelos (Regressor, Classifier)
│     │     • Transformadores (Imputer, Scaler, Encoder)
│     │     • Pipelines
│     │
│     └── Exemplo
│           • SimpleImputer → aprende média/mediana
│           • RandomForestRegressor → aprende padrões
│
├── 3. Interface fit / transform / predict
│     │
│     ├── fit()
│     │     • Treina o estimator
│     │     • Modelos aprendem padrões
│     │     • Transformadores aprendem parâmetros
│     │
│     ├── transform()
│     │     • Aplica transformação nos dados
│     │     • Usado por transformadores
│     │
│     ├── fit_transform()
│     │     • Combina fit + transform
│     │     • Usado em pré-processamento
│     │
│     └── predict()
│           • Usado apenas por modelos
│           • Gera previsões
│
└── 4. Pipelines
      │
      ├── Definição
      │     • Sequência organizada de transformações + modelo
      │     • Tudo vira um único estimator
      │
      ├── Vantagens
      │     • Evita data leakage
      │     • Código limpo e organizado
      │     • Reproduzível
      │     • Funciona com cross-validation
      │
      ├── Estrutura
      │     Pipeline(steps=[
      │         ('imputer', SimpleImputer()),
      │         ('scaler', StandardScaler()),
      │         ('model', RandomForestRegressor())
      │     ])
      │
      └── Comportamento
            • pipeline.fit() → treina tudo
            • pipeline.predict() → aplica transformações + modelo
```
